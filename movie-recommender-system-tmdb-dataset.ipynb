```python
#making string as a list

import ast

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head(1)
```

```python
#cast column
#we need first 3 dictionary. first 3 actors

movies['cast'][0]
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head(1)
```

```python
#crew column
#extract that dictionary where job=director

movies['crew'][0]
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(1)
```

```python
#getting list for each column
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
# removing spaces, because same name could be there but different person

def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head(1)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
#creating new dataframe with the required columns

new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
new_df['tags'][0]
```

```python
#lower case
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
new_df.head()
```

#Text Vectorizer

First add all the tags. Then take first 5000 most frequent words. Then go to each tag and see how many times those frequent words are coming in that particular tag. In this way a tag becomes a vector of dimention 5000.

Note - we will not take stop words like: are, of, and etc..

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors =  cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
#to check what are the 5000 most frequent words
cv.get_feature_names_out()
```

```python
cv.get_feature_names_out().tolist()
```

```python
#there are similar words like - 'abuse', 'abusive'.. 'accept', 'accepts' etc..
#to convert these into same words, we will use stemmer

import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
#again we need to run count vectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vectors =  cv.fit_transform(new_df['tags']).toarray()
cv.get_feature_names_out().tolist()
```

```python
#Now we will measure angular distance (cosine distance) between vectors. If it is less, vectors are similar, i.e., movies are similar
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
vectors.shape
```

```python
similarity.shape
```

So, now from 4806 vectors, each vector has calculated angular distance with each 4806 vectors

```python
similarity[0]
```

```python
#making tuples (index position, value)  using enumerate function. Its importance can be seen in the next step while creating function
list(enumerate(similarity[0]))
```

So, the similarity between one movie with itself will be 1. That's shown in the first value.

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(similarity[movie_index])),reverse=True,key = lambda x: x[1])[1:6]
    #key in the last is basically helps to sort on the basis of 2nd value in the tuple, i.e., distance value
    #[1:6] means taking the first 5 names
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
